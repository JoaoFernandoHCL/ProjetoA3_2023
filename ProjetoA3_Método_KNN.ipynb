```python
#importando a base utilizada no projeto
#https://www.kaggle.com/datasets/uom190346a/sleep-health-and-lifestyle-dataset
import pandas as pd
from google.colab import files
uploaded = files.upload()
df = pd.read_csv("Sleep_health_and_lifestyle_dataset.csv")
print(df)
```

```python
#Tratamento dos dados para ser utilizado no Método de IA escolhido

df = df.drop(["Daily Steps", "Person ID", "Occupation", "Heart Rate", "Gender"], axis = 1)
df = df.replace("Normal Weight", "Normal")
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
df["BMI Category"] = le.fit_transform(df["BMI Category"])
df['Sleep Disorder'] = le.fit_transform(df['Sleep Disorder'])
new = df['Blood Pressure'].str.split("/", n =1, expand = True)
media = []
for i in range(0, 374):
  media.append((int(new[0][i]) + int(new[1][i]))/2)
df["BloodPressure"] = media
df.drop(["Blood Pressure"], axis = 1, inplace= True)
x = df[["Sleep Duration", "Physical Activity Level", "Stress Level", "BMI Category", 'Sleep Disorder', 'BloodPressure']]
y = df["Quality of Sleep"]
```

```python
#Utilização do Método de IA escolhido
#Método KNN
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import numpy as np

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.4)

from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=6, metric='euclidean')
knn.fit(x_train, y_train)

y_pred = knn.predict(x_test)
```

```python
# Retorna a matriz de confusão mostrando os acertos e erros da nossa previsão
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test, y_pred))
from sklearn.metrics import classification_report
print('\n',classification_report(y_test, y_pred))
```
